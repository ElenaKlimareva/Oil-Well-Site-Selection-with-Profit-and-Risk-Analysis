# Oil Well Site Selection with Profit and Risk Analysis

**Project Description**

**Client:** an oil and gas company considering investment in new drilling operations  

The client is assessing three regions as potential locations for a new oil well. The available data include information on oil quality and reserve volumes from 100,000 deposits in each region. The objective is to determine where drilling would be economically viable while minimising financial risk.

To address this task, machine learning models are applied to estimate oil reserves at individual deposits, followed by an economic evaluation of expected drilling outcomes. Risk and profitability are further assessed using the Bootstrap method, enabling a probabilistic comparison of the regions.

The project delivers data-driven recommendations identifying the most promising region for investment, based on expected profit and associated risks.

---

**Project Goal**

**The goal of this project** is to identify the most suitable region for drilling a new oil well using machine learning–based reserve estimation combined with profit and risk analysis via the Bootstrap method.

## Model training and validation

**A linear regression model is trained for each region:**

### First region (geo_data_0)

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25
```

```python
X_0 = geo_data_0.drop(columns=['id', 'product'])
y_0 = geo_data_0['product']

X_train_0, X_valid_0, y_train_0, y_valid_0 = train_test_split(X_0, y_0, \
                                                            test_size=TEST_SIZE, \
                                                            random_state=RANDOM_STATE)
```

```python
num_columns = ['f0', 'f1', 'f2'] 
```

Since no clear linear relationship between the target variable and the features was observed in the `geo_data_0` and `geo_data_2` regions, polynomial features are added to capture potential non-linear relationships.

```python
num_pipe = Pipeline([
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('select', SelectFromModel(LinearRegression(), threshold='mean')),
    ('scaler', MinMaxScaler())
])


preprocessor = ColumnTransformer([
    ('num', num_pipe, num_columns)
])

pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
```

```python
param_grid = {
    'preprocessor__num__poly__degree': [2, 3, 4, 5],
    'preprocessor__num__select__threshold': ['mean', 'median', 0.1],
    'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler()]
    
}
```

```python
grid_search = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,                   
    scoring='neg_root_mean_squared_error',   
    cv=5,                                    
    n_jobs=-1,                                                             
)
```

```python
grid_search.fit(X_train_0, y_train_0)


print('The best-performing model and its parameters for geo_data_0:\n\n', grid_search.best_estimator_)
print('The best cross-validated RMSE:', -grid_search.best_score_.round(2))


y_valid_pred_0 = grid_search.predict(X_valid_0)

rmse_0 = mean_squared_error(y_valid_0, y_valid_pred_0, squared=False)
```

```python
X_valid_with_preds_0 = X_valid_0.copy()
X_valid_with_preds_0["prediction"] = y_valid_pred_0
X_valid_with_preds_0["target"] = y_valid_0.values

print('The model RMSE for the geo_data_0 dataset:', rmse_0.round(2))

print('The average predicted reserve volume for the geo_data_0 dataset:',\
      X_valid_with_preds_0['prediction'].mean().round(2))

print('The actual average reserve volume for the geo_data_0 dataset:', \
      X_valid_with_preds_0['target'].mean().round(2))
```

**Residual analysis:**

```python
def residuals(y_valid, y_valid_pred):
    residuals = y_valid - y_valid_pred
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7, 3.5))
    axes[0].hist(residuals)
    axes[0].set_title('Histogram of the residual distribution')
    axes[0].set_xlabel('Residuals')
    axes[1].scatter(y_valid_pred, residuals)
    axes[1].set_xlabel('Model predictions')
    axes[1].set_ylabel('Residuals')
    axes[1].set_title('Variance analysis')
    plt.tight_layout()
    plt.show()
```

```python
residuals(y_valid_0, y_valid_pred_0)
```

#### Conclusion on the model for the first region

To predict oil reserves in the first region, a **linear regression model** was trained. During model selection, the following steps were taken to improve performance:

- **Polynomial features** up to **degree 5** were introduced using `PolynomialFeatures` to capture non-linear relationships.

- **Feature selection** was performed using `SelectFromModel` with linear regression as the base estimator. Features with coefficients above the median importance threshold (`threshold='median'`) were retained.

- Features were **scaled** using `StandardScaler`.

- Model hyperparameters (polynomial degree, scaling method, and feature selection threshold) were tuned using **GridSearchCV** with **cross-validation**.

---

**Model results**

- **Best model configuration**:
  - `PolynomialFeatures(degree=5)`
  - `SelectFromModel(LinearRegression(), threshold='median')`
  - `StandardScaler()`

- **Best cross-validated RMSE**: **37.19**

- **RMSE on the validation set**: **37.15**

- **Average predicted reserve volume**: **92.68**

- **Actual average reserve volume**: **92.39**

Overall, the **average model error is approximately ±37.27 units**, which meets the client’s requirements. The model also accurately estimates the average reserve level, introducing almost no bias.

---

**Residual analysis**

- The **residual histogram** is close to a normal distribution, indicating balanced model performance.

- The **residual scatter plot** shows increasing error dispersion as predicted values grow. This may suggest that the model predicts higher reserve values with lower accuracy.

---

**Final remarks**

Despite its simplicity, the linear regression model demonstrates solid performance after the introduction of polynomial features and importance-based feature selection. However:

- The capabilities of a linear model may be close to their limit; further improvements in performance are likely to require more complex models (e.g. decision trees or gradient boosting), **if permitted**.

- Additional improvements could be achieved by **expanding the feature set** through feature engineering or by incorporating external information about wells and regions.

## Preparation for the profit calculation stage

### Sufficient reserve volume for break-even development

**The reserve volume required for break-even development of a new well is calculated and compared with the average reserve volume in each region**

**Task assumptions:**

The development budget for wells in a region is 10 billion roubles.

At current prices, one barrel generates 450 roubles in revenue. Since the reserve volume is measured in thousand barrels, the revenue per unit of product is 450,000 roubles.

### A function for profit calculation

```python
def revenue(wells, income_per_thousand=450_000, budget=10_000_000_000):
   
    top200_idx = wells['prediction'].nlargest(200).index
    
    top200_target = wells.loc[top200_idx, 'target']
    
    result = top200_target.sum() * income_per_thousand
    
    profit = result - budget
    
    return profit  
```

## Profit and risk estimation

```python
def bootstrap(X_valid_with_preds, n_bootstrap=1000, number_of_sq=500):
    
    state = np.random.RandomState(12345)
    values = []

    for _ in range(n_bootstrap):
        subsample = X_valid_with_preds.sample(
            n=number_of_sq,
            replace=True,
            random_state=state.randint(0, 1e6)
        )
        profit = revenue(subsample)  
        values.append(profit)

    values = pd.Series(values)
    mean = values.mean()
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    risk = (values < 0).mean()

    return values, mean, lower, upper, risk
```

```python
values_1, mean_1, lower_1, upper_1, risk_1 = bootstrap(X_valid_with_preds_0)

print(f'Average profit for the first region: {mean_1/1e9:.2f} billions.')
print(f'95% confidence interval for the first region: [{lower_1/1e9:.2f}, {upper_1/1e9:.2f}] billions.')
print(f'Risk of losses for the first region: {risk_1:.2%}.')
```

```python
def profit_distribution_hist(values, mean, lower, upper, risk, region):
    plt.hist(values/1e9, bins=50, edgecolor='black', alpha=0.7)

    plt.axvline(mean/1e9, color='red', linestyle='--', label=f'Mean = {mean/1e9:.2f}')
    plt.axvline(lower/1e9, color='green', linestyle='--', label=f'2.5% = {lower/1e9:.2f}')
    plt.axvline(upper/1e9, color='green', linestyle='--', label=f'97.5% = {upper/1e9:.2f}')
    plt.axvline(0, color='black', linestyle=':', label='Zero profit')

    plt.xlabel('Profit, billion roubles')
    plt.ylabel('Frequency')
    plt.title(f'Profit distribution ({region})')
    plt.legend(
        loc='upper right',
        fontsize=8
    )
    plt.show()
```
